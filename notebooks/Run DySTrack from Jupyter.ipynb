## Running DySTrack within a jupyter notebook

```python
# Imports

import os, sys

from dystrack.manager.manager import run_dystrack_manager
```

```python
# Load config file

sys.path.insert(0, r"..")
import run.run_lateral_line as run_ll
```

```python
# Further prep as needed

target_dir = "testrun_ipynb"

if not os.path.isdir(target_dir):
    os.mkdir(target_dir)
    
run_ll.analysis_kwargs["show"] = True
```

```python
# Run DySTrack

coordinates, stats_dict = run_dystrack_manager(
    target_dir, 
    run_ll.image_analysis_func,
    max_triggers=2,
    end_on_esc=True,  # This won't work here; use max_triggers or interrupt
    **run_ll.manager_kwargs, 
    img_kwargs=run_ll.analysis_kwargs,
)
```

```python
# Print resulting coordinates

for coords in coordinates:
    print(coords)
```

```python
# Print run statistics

for stat in stats_dict:
    print(f"{stat}: {stats_dict[stat]}")
```
